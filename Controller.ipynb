Consider the following simplified linear model of the drone
$$\begin{align}
    x_{n+1} = A
    x_n + B u_n
\end{align}$$

where the components of the state $x_n$ correspond to the horizontal position and velocity of the quadrotor, its vertical position and velocity and its orientation and angular velocity. The control vector
$u_n$ contains the forces produced by the rotors (our control inputs).
The module ```quadrotor.py``` defines the matrices $A$ and $B$ and code to generate a simulation of the drone.

````python
## what we need to do computation and display the drone
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt

import quadrotor
````

We want to generate a control input will move the drone towards the point $(-3,3)$ starting from $(0,0)$. In order to
do so, we define a cost that penalizes the distance to the goal at each time step while minimizing velocities and the amount of trust needed,
i.e. we want to solve the following optimal control problem
$$\begin{align}
&\min_{x_0, u_0, x_1, u_1, \cdots} \frac{1}{2}\sum_{n=0}^{N} (x_n - x_{desired})^T Q (x_n - x_{desired}) + u_n^T R u_n\\
\textrm{subject to}\ \ & x_{n+1} = A x_{n} + B u_n\\
& x_0 = [0,0,0,0,0,0]^T
\end{align}$$
where $x_{desired} = [-3,0,3,0,0,0]^T$

1. Write down the first-order optimality conditions (KKT conditions) for the problem
2. Write code to solve the problem for N=500 time steps (you will need to find diagonal matrices $Q>0$ and $R>0$ to create a nice movement) by solving the KKT system of linear equations. Do NOT inverse the KKT matrix, instead use the [NumPy solve function](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html) which is numerically more stable and efficient.
4. Show plots of all the states of the robot as a function of time
5. Show plots of the optimal control as a function of time

# write down your code here

Constant variables:

````python
nx = 6 # dimensions of x
nu = 2 # dimensions of u
N = 500 # time steps
nvars = N*nu + (N+1)*nx # number of variables
dt = 0.01 # time step size
````

Cost Values

````python
Q = np.eye(nx)
Q[0, 0] = 10e2 # x position penalty
Q[1, 1] = 10e1 # angular position penalty
Q[2, 2] = 10e2 # y position penalty
Q[3, 3] = 10e1 # x velocity penalty
Q[4, 4] = 10e1 # angular velocity penalty
Q[5, 5] = 10e1 # y velocity penalty
R = (1/10e3)*np.eye(nu)
````

Defining G, A, B, C, d matrices

````python
def getCostMatrix(Q, R):
    G = np.zeros((nvars, nvars))
    for i in range(N):
        G[i*(nx+nu):i*(nx+nu)+nx, i*(nx+nu):i*(nx+nu)+nx] = Q # populating diagnol spaces with Q
        G[i*(nx+nu)+nx:i*(nx+nu)+nx+nu, i*(nx+nu)+nx:i*(nx+nu)+nx+nu] = R # populating diagnol spaces with R
    G[N*(nx+nu):N*(nx+nu)+nx, N*(nx+nu):N*(nx+nu)+nx] = Q
    return G

G = getCostMatrix(Q, R)

C = np.zeros([nx*(N+1), nvars])
d = np.zeros([nx*(N+1)])

C[0:6, 0:6] = np.eye(nx)
d[0:6] = np.zeros(6) #initial position

# A and B are defined
A = quadrotor.A
B = quadrotor.B

for i in range(N): # populating C(constraints) matrix
    C[(i+1)*nx:(i+2)*nx, (i+1)*nx+i*nu:(i+1)*(nx+nu)] = B
    C[(i+1)*nx:(i+2)*nx, (i+1)*(nx+nu):(i+2)*nx+(i+1)*nu] = -np.eye(nx)
    C[(i+1)*nx:(i+2)*nx, i*(nx+nu):(i+1)*nx+i*nu] = A
````

Making the KKT matrix and solving for x

````python
nconstraints = C.shape[0]
KKT = np.zeros((nvars+nconstraints, nvars+nconstraints))

# KKT Matrix
KKT[0:nvars, 0:nvars] = G
KKT[nvars:, 0:nvars] = C
KKT[0:nvars, nvars:] = C.T

f = np.zeros((nvars+nconstraints,1))

g = np.zeros((nvars))
for i in range(N):
    g[i*(nx+nu):(i+1)*nx+i*nu] = -np.array([[-3, 0, 3, 0, 0, 0]])@Q
f[0:nvars,0] = -g
f[nvars:,0] = d

x = np.linalg.solve(KKT, f)
````

Show Animation:

````python
u = np.squeeze(np.array([x[6:nvars:8], x[7:nvars:8]]))

quadrotor.animate_robot(np.zeros((6,1)), u, goal=[-3,3])
````

````python
def plot_values(x_val):
    # Extract state and control variables
    xpos = x_val[0::8]
    theta = x_val[1::8]
    ypos = x_val[2::8]
    xvel = x_val[3::8]
    omega = x_val[4::8]
    yvel = x_val[5::8]
    u1 = x_val[6::8]
    u2 = x_val[7::8]
       
    # Print control statistics
    print(f"Control 1 - Min: {np.min(u1[:N]):.4f}, Max: {np.max(u1[:N]):.4f}")
    print(f"Control 2 - Min: {np.min(u2[:N]):.4f}, Max: {np.max(u2[:N]):.4f}")
    
    # Create time vector
    t = np.linspace(0, N, N)
    plt.figure(figsize=(9, 9))
    
    # State plots
    plt.subplot(8, 1, 1)
    plt.ylabel('x position')
    plt.plot(t, xpos[:N], 'x')
    
    plt.subplot(8, 1, 2)
    plt.ylabel('theta')
    plt.plot(t, theta[:N], 'x')
    
    plt.subplot(8, 1, 3)
    plt.ylabel('y position')
    plt.plot(t, ypos[:N], 'x')
    
    plt.subplot(8, 1, 4)
    plt.ylabel('x velocity')
    plt.plot(t, xvel[:N], 'x')
    
    plt.subplot(8, 1, 5)
    plt.ylabel('omega')
    plt.plot(t, omega[:N], 'x')
    
    plt.subplot(8, 1, 6)
    plt.ylabel('y velocity')
    plt.plot(t, yvel[:N], 'x')
    
    plt.subplot(8, 1, 7)
    plt.ylabel('control 1')
    plt.ylim([-10, 10])
    plt.plot(t, u1[:N], 'x')
    
    plt.subplot(8, 1, 8)
    plt.ylabel('control 2')
    plt.ylim([-10, 10])
    plt.plot(t, u2[:N], 'x')
````

Plot values:

# Initial Problem:
State vector, $$x = \begin{pmatrix} p_x \\ \theta \\ p_y \\ v_x \\ \omega \\ v_y \end{pmatrix}$$
Input vector, $$u = \begin{pmatrix} u_1 \\ u_2 \end{pmatrix}$$
Optimal Control Problem,
$$\begin{align}
& \min_{x_n, u_n} \sum_{n=0}^{N-1} \left( \frac{1}{2}x_n^T Q x_n + \frac{1}{2}u_n^T R u_n - \bar{x}_n^T Q x_n \right) + \frac{1}{2}x_N Q x_N + \bar{x}_N Q x_N\\
\textrm{subject to} \ & x_{n+1} = \begin{bmatrix}1 & \Delta t \\ 0 & 1 \end{bmatrix}x_n + \begin{bmatrix}0\\ \Delta t \end{bmatrix} u_n\\
& x_0 = x_{init}
\end{align}$$

where $Q,R \geq0$, and A and B taken from 'quadrotor.py' as given before.

The problem can be rewritten in matrix form as
$$\begin{align}
\min_{y} & \frac{1}{2} y^T G y + g^T y\\
\textrm{subject to} \ & C y = d
\end{align}$$

## KKT conditions
The Lagrangian of the problem is $L(y,\lambda) = \frac{1}{2} y^T G y + g^T y + \lambda^T (M y - p)$ and the KKT conditions for optimality are
$$ \begin{bmatrix} G & C^T \\ C & 0 \end{bmatrix} \begin{pmatrix} y \\ \lambda \end{pmatrix} = \begin{pmatrix} -g \\ d \end{pmatrix}$$

## Adding Inequality Constraints
$$\textrm{subject to} \ \begin{bmatrix} -5 \\ -5 \end{bmatrix} \leq u_n \leq \begin{bmatrix} 5 \\ 5\end{bmatrix}$$


````python
from qpsolvers import solve_qp

# creating bounds on the control and states
H = np.zeros((2*N*(nu +nx)+2*nx, nvars))
h = np.zeros((2*N*(nu + nx)+2*nx))

## feel free to play with these bounds
u_max = np.array([5, 5])
x_lim = np.array([5, 6, 5, 25, 6, 25])

for i in range(N+1):
    if i < N:
        H[(nu+nx)*i+nx:(nu+nx)*i+nx+nu,i*(nx+nu)+nx:i*(nx+nu)+nx+nu] = np.eye(nu)
        h[(nu+nx)*i+nx:(nu+nx)*i+nx+nu] = u_max
        H[(nu+nx)*i+nx+N*(nu+nx)+nx:(nu+nx)*i+nu+nx+N*(nu+nx)+nx,i*(nx+nu)+nx:i*(nx+nu)+nx+nu] = -np.eye(nu)
        h[(nu+nx)*i+nx+N*(nu+nx)+nx:(nu+nx)*i+nu+nx+N*(nu+nx)+nx] = u_max
        
    H[(nu+nx)*i:(nu+nx)*i+nx,i*(nx+nu):(i)*(nx+nu)+nx] = np.eye(nx)
    h[(nu+nx)*i:(nu+nx)*i+nx] = x_lim
    H[(nu+nx)*i+N*(nu+nx)+nx:(nu+nx)*i+nx+N*(nu+nx)+nx,i*(nx+nu):(i)*(nx+nu)+nx] = -np.eye(nx)
    h[(nu+nx)*i+N*(nu+nx)+nx:(nu+nx)*i+nx+N*(nu+nx)+nx] = x_lim

x_con = solve_qp(G, g, H, h, C, d, solver="cvxopt")
````

Show Animation:

````python
u = np.squeeze(np.array([x_con[6:nvars:8], x_con[7:nvars:8]]))

quadrotor.animate_robot(np.zeros((6,1)), u, goal=[-3,3])
````

Plot Values:

````python
plot_values(x_con)
````
